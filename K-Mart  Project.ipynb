# Qs 1 .What was the best month for sales ? How much was earned that month?

```python
df_filtered["Quantity Ordered"]=\
df_filtered["Quantity Ordered"].\
astype("float")
df_filtered["Price Each"]=\
df_filtered["Price Each"].\
astype("float")       
```

```python
df_filtered.info()
```

```python
df_filtered["Price Each"].nunique
```

```python
df_filtered["sales"]=df_filtered["Quantity Ordered"]*df_filtered["Price Each"]
```

```python
df_filtered.head()
```

```python
df_filtered["Month"].unique()
```

```python
df_filtered.info()
```

```python
df=df_filtered.groupby(['Month'])\
['sales'].sum().sort_values(ascending=False)
df=df.to_frame()
```

```python
df.info
```

```python
df["sales"]
```

```python
df.columns
```

```python
month=["dec","oct","apr","nov",
       "may","mar","july","june"
       ,"aug",'feb',"sep","jan"] 
df["month"]=month
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)

sns.barplot(x = "sales",
            y = "month",
            data = df)
plt.title("Month wise Sale")
plt.show()
```

# Qs 2 . Which city had the highest number of sales ?

```python
df_filtered.info()
```

```python
dftemp = df_filtered
list_city = []
for i in dftemp['Purchase Address']:
    list_city.append(i.split(",")[1])
dftemp['City'] = list_city
dftemp.head()
```

```python
df_city=df_filtered.groupby(["City"])\
['sales'].sum().sort_values(ascending=False)
df_city=df_city.to_frame()
```

```python
df_city
```

```python
dftemp["City"].nunique()
```

```python
city=["San Francisco","Los Angeles",
      "New York City","Boston","Atlanta",
      "Dallas","Seattle","Portland","Austin"]
```

```python
df_city["city"]=city
```

```python
df_city
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.barplot(x = "sales",
            y = "city",
            data = df_city)
plt.title("City wise Sales")
plt.show()
```
